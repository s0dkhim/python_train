## Gmarket에 있는 베스트셀러 200품목 불러 오기
http://corners.gmarket.co.kr/Bestsellers

```python
from selenium import webdriver
import pandas as pd
```

```python

driver = webdriver.Chrome()
driver.get("http://corners.gmarket.co.kr/Bestsellers")


best_seller = driver.find_elements_by_css_selector("#gBestWrap .best-list")
```

```python
len(best_seller)
```

```python
def get_Gmarket_bestsellers(number=10):
    
    def to_int(price):
        return int(price.replace(",", "").replace("원", ""))
    
    driver = webdriver.Chrome()
    driver.get("http://corners.gmarket.co.kr/Bestsellers")
    
    
    best_seller = driver.find_elements_by_css_selector("#gBestWrap .best-list")[1]
    items = best_seller.find_elements_by_css_selector('li')[:number]

    
    df = pd.DataFrame(columns = ['순위', '이미지링크', '제품링크', '제품명', '정가', '판가'])
    
    for item in items:
        
        idx = item.find_element_by_css_selector('p').text
        img_link = item.find_element_by_css_selector('img').get_attribute('src')
        goods_link = item.find_element_by_css_selector('.itemname').get_attribute('href')
        name = item.find_element_by_css_selector('.itemname').text
        o_price = item.find_element_by_css_selector('.item_price .o-price').text
        s_price = item.find_element_by_css_selector('.item_price .s-price').text.split(' ')[0]
        o_price = o_price if o_price else s_price
        
        data = {'순위': idx,
                '이미지링크': img_link,
                '제품링크': goods_link,
                '제품명': name,
                '정가': to_int(o_price),
                '판가': to_int(s_price), }
        
        df.loc[len(df)] = data
        
    driver.close
    
    return df
    
```

```python
%%time
result = get_Gmarket_bestsellers(200)
result.tail()
```

## 셀레니움을 이용하여 2018년 2월 20일 네이버 IT/과학 기사의 제목을 리스트로 크롤링
http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105

```python
article_list = []

def get_article(page):
    driver = webdriver.Chrome()

    driver.get('http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=2018-02-20 00:00:00&page='+str(page))
    
#     driver.get("http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105")
#     driver.find_element_by_css_selector("#pagenavi_day a").click()
    
    
    articles = driver.find_elements_by_css_selector('#section_body li')
    
    for article in articles:
        title = article.find_element_by_css_selector('dt:not(.photo) > a').text
        
        article_list.append(title)
    
    driver.close()
    return article_list
```

```python
%%time
for page in range(1, 8+1):
    get_article(page)
```

## threading을 활용하면 훨씬 빠르게 할 수 있음.

```python
import threading

for page in range(1,9):
    th = threading.Thread(target=get_article, args=(page,))
    th.start()
    
article_list
```
