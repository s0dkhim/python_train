```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import sympy
import forecastio
import requests
import time
from selenium import webdriver
import re
```

```python

driver = webdriver.Chrome()
driver.get('https://www.premierleague.com/tables')

main_content = driver.find_element_by_css_selector('#mainContent')
Contents = main_content.find_elements_by_css_selector('.tableBodyContainer tr:not(.expandable)')

df = pd.DataFrame(columns = ['Position', 'Club', 'Played', 'Won', 'Drawn', 'Lost', 'GF', 'GA', 'GD', 'Points'])

for i in range(0, 0+20):
    Position = Contents[i].find_elements_by_css_selector('#Tooltip')[0].text
    Club = Contents[i].find_elements_by_css_selector('a')[0].text
    Played = Contents[i].find_elements_by_css_selector('td')[3].text
    Won = Contents[i].find_elements_by_css_selector('td')[4].text
    Drawn = Contents[i].find_elements_by_css_selector('td')[5].text
    Lost = Contents[i].find_elements_by_css_selector('td')[6].text
    GF = Contents[i].find_elements_by_css_selector('td')[7].text
    GA = Contents[i].find_elements_by_css_selector('td')[8].text
    GD = Contents[i].find_elements_by_css_selector('td')[9].text
    Points = Contents[i].find_elements_by_css_selector('td')[10].text
    
    data = {'Position': Position,
            'Club': Club,
            'Played': Played,
            'Won': Won,
            'Drawn': Drawn,
            'Lost': Lost,
            'GF': GF,
            'GA': GA,
            'GD': GD,
            'Points': Points, }
    
    df.loc[len(df)] = data
    
    driver.close

df
```
