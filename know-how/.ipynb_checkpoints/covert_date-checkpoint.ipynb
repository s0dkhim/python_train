```python
import pandas as pd
```

```python
# step1: raed csv file
hol = pd.read_table('holiday_names.txt', sep='\s+', names = ['year', 'month', 'day', 'name'])
```

```python
hol.month.unique()
```

```python
# step2 : create dictionary to match month

dic = {'Jan': 1,
       'Feb': 2,
       'Mar': 3,
       'Apr': 4,
        'May': 5,
        'Jun': 6,
        'Jul': 7,
        'Aug': 8,
         'Sep': 9,
         'Oct': 10,
          'Nov': 11,
          'Dec': 12, }
```

```python
# step3: Use map to apply dic to the month column of hol  
hol['monthnumber'] = hol.month.map(dic)
```

```python
# step4: pd.to_datetime 기능을 활용하기 위해 기존 month 컬럼의 이름을 month name으로 변경하고 monthnumber 컬럼명을 month로 변경.
hol.columns = ['year', 'month name', 'day', 'name', 'month']


hol = hol.assign(date = pd.to_datetime(hol[['year', 'month', 'day']]))
hol.info()
hol
```
